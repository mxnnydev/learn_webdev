### üîπ PART 1: Variables ‚Äì `const` vs `let`

* Use `const` when the **value never needs to be reassigned** (default).
* Use `let` when the **value might change** (like a counter or user input).

```js
const appName = "FitSnap AI";
let userCount = 0;

userCount += 1;
```

üß† **Why this matters:** UIs are full of dynamic data ‚Äî `let` lets you change it.

### üîπ PART 3: Arrays ‚Äì Your UI Data Engine

Arrays store multiple values. You‚Äôll loop over them to create cards or lists.

```js
const users = ["Alice", "Bob", "Charlie"];

users.forEach(user => {
  console.log(user); // for UI: createCard(user)
});
```

#### Key methods:

* `.push()` ‚Äî add to end
* `.pop()` ‚Äî remove from end
* `.map()` ‚Äî create new transformed array

```js
const cards = users.map(name => `<div>${name}</div>`);
```

üí° `map()` is a game-changer when you render dynamic components in React.

### üîπ PART 4: Objects ‚Äì For Real-World Data

Objects are used everywhere ‚Äî especially in APIs.

```js
const user = {
  name: "Emmanuel",
  age: 25,
  skills: ["Python", "JS"]
};

console.log(user.name); // Emmanuel
console.log(user.skills[0]); // Python
```

‚úÖ Access nested data:

```js
const data = {
  profile: {
    username: "builderboss",
    stats: { followers: 1200 }
  }
};

console.log(data.profile.stats.followers); // 1200
```

### üîπ PART 5: Functions & Arrow Functions

Functions let you organize and reuse logic.

```js
function greet(name) {
  return `Hello, ${name}`;
}
```

Arrow functions ‚Äî cleaner, faster syntax:

```js
const greet = name => `Hello, ${name}`;
```

üí° Use them inside `map`, `forEach`, `filter`, etc.
