### 🛠️ Phase 1: Core HTML Building Blocks (1-2 Days)

**Goal**: Understand the structure and syntax of HTML to create basic pages.

- [ ]  **HTML structure**: `<html>`, `<head>`, `<body>` — what they do
- [ ]  **Text elements**: `<h1>` to `<h6>`, `<p>`, `<b>`, `<i>`, `<strong>`, `<em>`
- [ ]  **Links & Navigation**: `<a href="">` (internal vs external links)
- [ ]  **Lists**: `<ul>`, `<ol>`, `<li>`
- [ ]  **Images**: `<img src="">` with `alt` text
- [ ]  **Breaks and Dividers**: `<br>`, `<hr>`

🧪 **Mini Project**: Create a simple landing page for your SaaS idea.

---

### 🎯 Phase 2: Forms & User Input (2 Days)

**Goal**: Collect user input (signup, login, feedback).

- [ ]  `<form>`, `action`, `method`
- [ ]  Input fields: `<input type="text">`, `<email>`, `<password>`, `<submit>`
- [ ]  Labels and placeholders
- [ ]  Buttons: `<button>`, `<input type="submit">`
- [ ]  Text areas: `<textarea>`

🧪 **Mini Project**: Build a working signup form for your app (doesn’t need backend yet).

---

### 🧩 Phase 3: Layout & Containers (2 Days)

**Goal**: Organize content visually and semantically.

- [ ]  `<div>`, `<span>` – wrappers
- [ ]  `<section>`, `<header>`, `<footer>`, `<nav>` – HTML5 semantic elements
- [ ]  `<table>` basics if you're displaying simple data

🧪 **Mini Project**: Create a dashboard layout for your SaaS tool (static HTML).

---

### 🚀 Phase 4: Launch Prep (1 Day)

**Goal**: Learn what you need to go live.

- [ ]  How to structure files: `index.html`, `assets/`, `styles/`
- [ ]  Hosting with GitHub Pages or Netlify
- [ ]  SEO Basics: `<title>`, `<meta>`, alt tags, `<h1>` structure
- [ ]  Accessibility basics (labels, alt text, heading order)

🧪 **Mini Project**: Deploy your landing page live.
